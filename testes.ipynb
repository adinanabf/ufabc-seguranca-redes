```python
# list_webcams.py
# Scans and lists available webcam indices using OpenCV.
# Usage in Jupyter: put this cell at index 0 and run.

import cv2
import time
import sys

def list_cameras(max_devices=10, read_test=True, wait=0.1):
    """
    Return a list of available camera indices.
    - max_devices: how many indices to probe (0..max_devices-1)
    - read_test: if True, attempt to read one frame to verify the device works
    - wait: seconds to wait after opening before attempting read
    """
    found = []
    # choose backend hint for faster open on Windows/Linux if available
    backend = None
    if sys.platform.startswith("win"):
        backend = cv2.CAP_DSHOW
    elif sys.platform.startswith("linux"):
        backend = cv2.CAP_V4L2

    for i in range(max_devices):
        if backend is not None:
            cap = cv2.VideoCapture(i, backend)
        else:
            cap = cv2.VideoCapture(i)
        time.sleep(wait)
        if not cap.isOpened():
            cap.release()
            continue
        ok = True
        if read_test:
            ok, _ = cap.read()
        cap.release()
        if ok:
            found.append(i)
    return found

if __name__ == "__main__":
    cams = list_cameras(max_devices=16)
    if cams:
        print("Found camera indices:", cams)
    else:
        print("No cameras found.")
```
